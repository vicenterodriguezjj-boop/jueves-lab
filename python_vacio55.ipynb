#### **2.2 SETS**

```python
################### 1.CREACIÓN DE SETS #####################

# 1. set vacío
set_vacio = set()

# 2. Inicializar un set con elementos
frutas = {'manzana', 'banana', 'cereza'}
print(frutas)


################### 2.INSERTAR Y ACTUALIZAR #####################

# Añadir elementos
frutas.add('naranja') # Añade un elemento si no existe
print(frutas)

# Actualizar con otro iterable (añade todos los elementos del iterable)
frutas.update(['uva', 'kiwi'])
print(frutas)


################### 3.BORRADOS #####################

# Eliminar un elemento (lanza error si no existe)
frutas.remove('banana')
print(frutas)

# Eliminar un elemento (no lanza error si no existe)
frutas.discard('mango')
print(frutas)

# Eliminar y retornar un elemento arbitrario
elemento_eliminado_set = frutas.pop()
print(f"Elemento eliminado: {elemento_eliminado_set}, Set actual: {frutas}")

# Vaciar el set completamente
frutas.clear()
print(frutas)

# Eliminar el set completamente (variable)
# del frutas


################### 4.OPERACIONES DE SETS #####################

set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# Unión (elementos en A o en B)
union_sets = set_a.union(set_b)
print(f"Unión: {union_sets}") # Output: {1, 2, 3, 4, 5, 6}

# Intersección (elementos comunes en A y B)
interseccion_sets = set_a.intersection(set_b)
print(f"Intersección: {interseccion_sets}") # Output: {3, 4}

# Diferencia (elementos en A pero no en B)
diferencia_sets = set_a.difference(set_b)
print(f"Diferencia (A-B): {diferencia_sets}") # Output: {1, 2}

# Diferencia simétrica (elementos en A o B pero no en ambos)
diferencia_simetrica_sets = set_a.symmetric_difference(set_b)
print(f"Diferencia simétrica: {diferencia_simetrica_sets}") # Output: {1, 2, 5, 6}

# Subconjunto (is_subset)
set_c = {1, 2}
print(f"¿C es subconjunto de A?: {set_c.issubset(set_a)}") # Output: True

# Superconjunto (is_superset)
print(f"¿A es superconjunto de C?: {set_a.issuperset(set_c)}") # Output: True


################### 5.ACCESO A LOS SETS #####################

# No hay una forma de acceder de forma indexada
# Para verificar la existencia de un elemento:
print(f"¿'manzana' está en frutas?: {'manzana' in frutas}")

# Iterar sobre un set
for fruta in {'manzana', 'banana'}:
    print(fruta)
```

#### **2.3 TUPLAS**

```python
################### 1.CREACIÓN DE TUPLAS #####################

# 1. Tupla vacía
tupla_vacia = ()
tupla_vacia_constructor = tuple()

# 2. Tupla con elementos
coordenadas = (10, 20)
print(f"Coordenadas: {coordenadas}")

# 3. Tupla con un solo elemento (¡nota la coma!)
solo_elemento = ('python',)
print(f"Tupla de un solo elemento: {solo_elemento}")

# 4. Empaquetado de tuplas
valores = 1, 2, 3
print(f"Tupla por empaquetado: {valores}")

################### 2.INSERTAR Y ACTUALIZAR #####################

# Las tuplas son inmutables, lo que significa que no se pueden insertar, añadir o eliminar elementos
# directamente una vez creadas. Si necesitas modificar una tupla, la práctica común es crear una nueva tupla.

# Ejemplo de 'modificación' (creando una nueva tupla)
lista_temporal = list(coordenadas)
lista_temporal.append(30)
nuevas_coordenadas = tuple(lista_temporal)
print(f"Nuevas coordenadas (simulando adición): {nuevas_coordenadas}")

################### 3.BORRADOS #####################

# Las tuplas no soportan la eliminación directa de elementos debido a su inmutabilidad.
# Para 'eliminar' un elemento, se crea una nueva tupla con los elementos deseados.

# Ejemplo de 'eliminación' (creando una nueva tupla sin un elemento)
otra_tupla = (1, 2, 3, 4, 5)
tupla_sin_tres = otra_tupla[:2] + otra_tupla[3:]
print(f"Tupla original: {otra_tupla}, Tupla sin el 3: {tupla_sin_tres}")

# Eliminar la tupla completamente (la variable)
# del coordenadas
# print(coordenadas) # Esto daría un NameError

################### 4.OPERACIONES DE TUPLAS #####################

tupla_a = (1, 2, 3)
tupla_b = (4, 5, 6)

# Concatenación
tupla_concatenada = tupla_a + tupla_b
print(f"Concatenación: {tupla_concatenada}")

# Repetición
tupla_repetida = tupla_a * 2
print(f"Repetición: {tupla_repetida}")

# Longitud
print(f"Longitud de tupla_a: {len(tupla_a)}")

# Pertenencia
print(f"¿Está 2 en tupla_a?: {2 in tupla_a}")

################### 5.ACCESO A LOS TUPLAS #####################

# Acceso de forma indexada
mi_tupla = ('a', 'b', 'c', 'd')
print(f"Primer elemento: {mi_tupla[0]}")
print(f"Último elemento: {mi_tupla[-1]}")

# Acceso con intervalos de elementos (slicing)
print(f"Elementos del índice 1 al 2: {mi_tupla[1:3]}")
print(f"Desde el segundo elemento hasta el final: {mi_tupla[1:]}")
print(f"Desde el principio hasta el tercer elemento: {mi_tupla[:3]}")

# Acceso con bucle for
print("Iterando sobre la tupla:")
for elemento in mi_tupla:
    print(elemento)
```

### **4. BUCLES**

```python
############################# BUCLES FOR IN #####################################
#Sintaxis:
#For elemento in elementos:
#   Acciones

# Ejemplo 1: Iterar sobre una lista
frutas = ["manzana", "banana", "cereza"]
print("Ejemplo For 1 (lista):")
for fruta in frutas:
    print(fruta)

# Ejemplo 2: Iterar sobre un string
nombre = "Python"
print("\nEjemplo For 2 (string):")
for letra in nombre:
    print(letra)

# Ejemplo 3: Iterar con range()
print("\nEjemplo For 3 (range):")
for i in range(5): # Genera números del 0 al 4
    print(i)

# Ejemplo 4: Iterar sobre las claves de un diccionario
diccionario = {"a": 1, "b": 2}
print("\nEjemplo For 4 (claves diccionario):")
for clave in diccionario:
    print(clave)


############################# BUCLES WHILE #####################################
#Sintaxis:
#While condicion:
#   Acciones

# Ejemplo 1: Bucle while simple
contador = 0
print("\nEjemplo While 1:")
while contador < 3:
    print(f"El contador es {contador}")
    contador += 1

# Ejemplo 2: Bucle while con condición booleana
es_verdadero = True
numero = 0
print("\nEjemplo While 2:")
while es_verdadero:
    print(f"El número es {numero}")
    numero += 1
    if numero == 2:
        es_verdadero = False # Cambia la condición para salir del bucle


############################# SENTENCIAS BREAK, PASS Y CONTINUE #####################################

#Break: Detiene la ejecución del bucle.
print("\nEjemplo Break:")
for i in range(5):
    if i == 3:
        break # Sale del bucle cuando i es 3
    print(i)

#Pass: Se utiliza para dejar vacío un trozo de código que todavía no queremos especificar.
print("\nEjemplo Pass:")
for i in range(3):
    if i == 1:
        pass # No hace nada, simplemente pasa a la siguiente instrucción
    print(i)

#Continue: Pasa a la siguiente iteración del bucle.
print("\nEjemplo Continue:")
for i in range(5):
    if i == 2:
        continue # Omite el resto del código en esta iteración cuando i es 2
    print(i)
```

```python
# Sintaxis: def nombre_funcion(parametro1 : Tipo1, parametro2: Tipo2) -> TipoDatoRetorno:
#               Acciones
#               return Valor (Opcional)

############################# FUNCIÓN SIN PARÁMETROS #####################################
def saludar():
    print("¡Hola, mundo!")

saludar()

############################# FUNCIÓN CON PARÁMETROS #####################################

# Python es un lenguaje con tipado dinámico, es decir, no hace falta indicar que tipo de datos va a recibir o devolver una función ya que Python
# va a inferir los tipo de datos en tiempo de ejecución. Es muy recomendable indicar los tipos de datos para aumentar la legibilidad del código.
# Es más, aunque tu le indiques el tipo de parámetros y luego le pasas otro tipo de datos NO va a dar un fallo.
def sumar(a: int, b: int) -> int:
    return a + b

resultado_suma = sumar(5, 3)
print(f"La suma es: {resultado_suma}")

############################# FUNCIÓN CON PARÁMETROS CON VALORES POR DEFECTO #####################################
def saludar_personalizado(nombre: str, mensaje: str = "¡Hola!") -> None:
    print(f"{mensaje} {nombre}!")

saludar_personalizado("Ana")
saludar_personalizado("Pedro", "¿Qué tal,")

############################# FUNCIÓNES QUE DEVUELVE MÁS DE UN VALOR #####################################
def operaciones_matematicas(num1: int, num2: int) -> tuple:
    suma = num1 + num2
    resta = num1 - num2
    return suma, resta

res_suma, res_resta = operaciones_matematicas(10, 4)
print(f"Suma: {res_suma}, Resta: {res_resta}")
```

### **6. PROGRAMACIÓN ORIENTADA A OBJETOS**

```python

# Sintaxis:

# class NombreClase():
#   Atributos
#   def __init__(self):
#       Inicializar atributos (constructor)
#   Setters and getters
#   Metodos de clase
#

# Implementación de la clase mascota para un videojuego de animales.

class Mascota:
    ############################# ATRIBUTOS DE LA CLASE #####################################
    def __init__(self, nombre: str, especie: str, salud: int = 100, nivel: int = 1):
        self.__nombre = nombre  # Atributo privado
        self.__especie = especie
        self.__salud = salud
        self.__nivel = nivel

    ############################# CONSTRUCTOR DE LA CLASE #####################################
    # El constructor (__init__) se define arriba junto a los atributos.

    ############################# METODOS GETTERS AND SETTERS #####################################

    # Getters
    def get_nombre(self) -> str:
        return self.__nombre

    def get_especie(self) -> str:
        return self.__especie

    def get_salud(self) -> int:
        return self.__salud

    def get_nivel(self) -> int:
        return self.__nivel

    # Setters
    def set_nombre(self, nombre: str) -> None:
        self.__nombre = nombre

    def set_salud(self, salud: int) -> None:
        if 0 <= salud <= 100: # Validar que la salud esté entre 0 y 100
            self.__salud = salud
        else:
            print("La salud debe estar entre 0 y 100.")

    def set_nivel(self, nivel: int) -> None:
        if nivel > 0:
            self.__nivel = nivel
        else:
            print("El nivel debe ser mayor que 0.")

    ############################# METODOS DE CLASE #####################################

    # Método que sirve para aumentar la salud a tu mascota.
    def curar(self, cantidad: int) -> None:
        self.set_salud(min(100, self.__salud + cantidad)) # Asegura que la salud no exceda 100
        print(f"{self.__nombre} ha sido curado. Salud actual: {self.__salud}")

    # Método que sirve para aumentar de nivel a tu mascota en función del torneo que haya ganado.
    def subir_nivel(self, tipo_torneo: str) -> None:
        if tipo_torneo == "principiante":
            self.set_nivel(self.__nivel + 1)
        elif tipo_torneo == "intermedio":
            self.set_nivel(self.__nivel + 2)
        elif tipo_torneo == "experto":
            self.set_nivel(self.__nivel + 3)
        else:
            print("Tipo de torneo no reconocido.")
        print(f"{self.__nombre} ha subido de nivel. Nivel actual: {self.__nivel}")

# Ejemplo de uso:
# mi_mascota = Mascota("Fido", "Perro")
# print(f"Nombre: {mi_mascota.get_nombre()}, Especie: {mi_mascota.get_especie()}, Salud: {mi_mascota.get_salud()}, Nivel: {mi_mascota.get_nivel()}")

# mi_mascota.curar(10)
# mi_mascota.subir_nivel("principiante")

# mi_mascota.set_salud(50)
# print(f"Salud de {mi_mascota.get_nombre()}: {mi_mascota.get_salud()}")
```
