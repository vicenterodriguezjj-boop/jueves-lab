# Actividad C. Limpieza y tratamiento de errores (25 puntos)

Crea `df_limpio` y realiza:
1. Limpieza de texto (`campus`, `modalidad`)
2. Normalización de etiquetas (ej. `slp centro`, `SLP Cntro`, `escolar`, `MIXTA`)
3. Eliminar duplicados por `id_estudiante`
4. Corregir rangos en `promedio` y `asistencia_pct` (0–100)
5. Corregir `creditos_aprobados <= creditos_inscritos`
6. Imputar faltantes (`edad`, `promedio`, `asistencia_pct`) con mediana

```python
# --- Respuesta del estudiante (C1) ---
# Crear df_limpio y realizar limpieza
df_limpio = df.copy()

# 1. Limpieza de texto y normalización para 'campus'
campus_mapping = {
    'slp centro': 'SLP Centro',
    'slp cntro': 'SLP Centro',
    'slp norte': 'SLP Norte',
    'matehuala': 'Matehuala'
}
df_limpio['campus'] = df_limpio['campus'].str.lower().str.strip().replace(campus_mapping)

# 2. Limpieza de texto y normalización para 'modalidad'
modalidad_mapping = {
    'escolarizada': 'Escolarizada',
    'escolar': 'Escolarizada',
    'mixta': 'Mixta',
    'mixta ': 'Mixta'
}
df_limpio['modalidad'] = df_limpio['modalidad'].str.lower().str.strip().replace(modalidad_mapping)

# 3. Eliminar duplicados por id_estudiante
df_limpio.drop_duplicates(subset=['id_estudiante'], keep='first', inplace=True)

# 4. Corregir rangos en promedio y asistencia_pct (0-100)
df_limpio['promedio'] = df_limpio['promedio'].clip(0, 100)
df_limpio['asistencia_pct'] = df_limpio['asistencia_pct'].clip(0, 100)

# 5. Corregir creditos_aprobados <= creditos_inscritos
df_limpio['creditos_aprobados'] = np.minimum(df_limpio['creditos_aprobados'], df_limpio['creditos_inscritos'])

# 6. Imputar faltantes (edad, promedio, asistencia_pct) con mediana
median_edad = df_limpio['edad'].median()
median_promedio = df_limpio['promedio'].median()
median_asistencia_pct = df_limpio['asistencia_pct'].median()

df_limpio['edad'] = df_limpio['edad'].fillna(median_edad)
df_limpio['promedio'] = df_limpio['promedio'].fillna(median_promedio)
df_limpio['asistencia_pct'] = df_limpio['asistencia_pct'].fillna(median_asistencia_pct)
```

```python
# --- Respuesta del estudiante (C2) ---
# Mostrar df_limpio.head()
display(df_limpio.head())
```

```python
# --- Respuesta del estudiante (C3) ---
# Verificación posterior de calidad de datos

print("Verificación de Faltantes en df_limpio:")
display(df_limpio.isnull().sum())

print("\nVerificación de Duplicados por 'id_estudiante' en df_limpio:")
duplicated_ids_limpio = df_limpio[df_limpio.duplicated(subset=['id_estudiante'], keep=False)]
display(duplicated_ids_limpio)
print(f"Total de filas duplicadas (incluyendo la primera aparición) después de limpieza: {len(duplicated_ids_limpio)}")

print("\nVerificación de Valores únicos para 'campus' en df_limpio (después de limpieza):")
display(df_limpio['campus'].unique())

print("\nVerificación de Valores únicos para 'modalidad' en df_limpio (después de limpieza):")
display(df_limpio['modalidad'].unique())

print("\nVerificación de 'promedio' fuera de rango (0-100) en df_limpio:")
display(df_limpio[(df_limpio['promedio'] < 0) | (df_limpio['promedio'] > 100)])

print("\nVerificación de 'asistencia_pct' fuera de rango (0-100) en df_limpio:")
display(df_limpio[(df_limpio['asistencia_pct'] < 0) | (df_limpio['asistencia_pct'] > 100)])

print("\nVerificación de error lógico: 'creditos_aprobados' > 'creditos_inscritos' en df_limpio:")
display(df_limpio[df_limpio['creditos_aprobados'] > df_limpio['creditos_inscritos']])
```

**Respuesta:** Conviene limpiar sobre una copia (`df_limpio`) para preservar el DataFrame original (`df`) en su estado crudo. Esto es crucial por varias razones:

# Actividad D. Filtración y transformación (20 puntos)

Con `df_limpio`:
1. Filtra `promedio < 80`
2. Filtra `campus == 'SLP Centro'` y `modalidad == 'Escolarizada'`
3. Crea `tasa_aprobacion = creditos_aprobados / creditos_inscritos`
4. Crea `riesgo_academico`:
   - Alto: `promedio < 70` o `asistencia_pct < 75`
   - Medio: `promedio < 80`
   - Bajo: otro caso
5. (Opcional) One-hot encoding de `modalidad`

```python
# --- Respuesta del estudiante (D1) ---
# Filtros solicitados

# 1. Filtra promedio < 80
df_promedio_bajo = df_limpio[df_limpio['promedio'] < 80]
print("Estudiantes con promedio < 80:")
display(df_promedio_bajo)

# 2. Filtra campus == 'SLP Centro' y modalidad == 'Escolarizada'
df_slp_escolarizada = df_limpio[(df_limpio['campus'] == 'SLP Centro') & (df_limpio['modalidad'] == 'Escolarizada')]
print("\nEstudiantes de 'SLP Centro' en modalidad 'Escolarizada':")
display(df_slp_escolarizada)
```

```python
# --- Respuesta del estudiante (D2) ---
# Crear tasa_aprobacion y riesgo_academico

# 3. Crea tasa_aprobacion = creditos_aprobados / creditos_inscritos
df_limpio['tasa_aprobacion'] = df_limpio['creditos_aprobados'] / df_limpio['creditos_inscritos']

# 4. Crea riesgo_academico:
#    - Alto: promedio < 70 o asistencia_pct < 75
#    - Medio: promedio < 80
#    - Bajo: otro caso
conditions = [
    (df_limpio['promedio'] < 70) | (df_limpio['asistencia_pct'] < 75),
    (df_limpio['promedio'] < 80)
]
choices = ['Alto', 'Medio']
df_limpio['riesgo_academico'] = np.select(conditions, choices, default='Bajo')

# 5. (Opcional) One-hot encoding de modalidad
# Aunque es opcional, lo incluiremos para mostrar cómo se hace.
df_limpio = pd.get_dummies(df_limpio, columns=['modalidad'], prefix='modalidad', dtype=int)
```

```python
# --- Respuesta del estudiante (D3) ---
# Mostrar evidencia de transformación
print("DataFrame después de crear 'tasa_aprobacion' y 'riesgo_academico' y One-Hot Encoding de 'modalidad':")
display(df_limpio.head())

print("\nDistribución de 'riesgo_academico':")
display(df_limpio['riesgo_academico'].value_counts())

print("\nColumnas del DataFrame después del One-Hot Encoding de 'modalidad':")
display(df_limpio.columns)
```

### Respuesta breve del estudiante (D4)
¿Cuál es la diferencia entre **filtrar** y **transformar** datos?

**Respuesta:**

-   **Filtración (Filtering):** La filtración de datos implica seleccionar un subconjunto de filas (observaciones) de un DataFrame que cumplen con una o más condiciones específicas. Esencialmente, se eliminan las filas que no son relevantes para el análisis actual, pero las columnas y la estructura de las filas restantes no cambian. El objetivo es reducir el volumen de datos o enfocarse en un grupo particular de observaciones.
    *   **Ejemplo:** Seleccionar solo a los estudiantes con un promedio menor a 80, o aquellos que pertenecen a un campus específico.

-   **Transformación (Transformation):** La transformación de datos implica modificar la estructura o los valores de las columnas (variables) existentes, o crear nuevas columnas a partir de las existentes. No se eliminan filas, sino que se alteran los datos para hacerlos más adecuados para el análisis, para normalizarlos, para derivar nuevas características, o para cambiar su tipo de dato.
    *   **Ejemplo:** Calcular la `tasa_aprobacion` (una nueva columna), categorizar el `riesgo_academico` (modificar valores o crear una nueva columna), o aplicar One-Hot Encoding a una columna categórica como `modalidad` (crear nuevas columnas binarias).
