```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Preparación de los datos (tu código original)
n = 120
X = np.linspace(0, 100, n)
y_real = 2.0 * -X
y = y_real + np.random.normal(0, 25, n)

# 2. Cálculo de los residuales
residuales = y - y_real

# 3. Generación del Histograma de Residuales
fig = plt.figure(figsize = (6, 4))
ax = plt.gca()

# Usamos los residuales calculados en lugar de la muestra aleatoria
ax.hist(residuales, bins = 20, color = "skyblue", edgecolor = "white", alpha = 0.8)

ax.set_title("Histograma de los Residuales")
ax.set_xlabel("Error (Residual)")
ax.set_ylabel("Frecuencia")

ax.grid(True, axis = 'y', linestyle = "--", alpha = 0.4)

plt.show()
```

```python

# 1. Configuración de parámetros
n = 1000  # Aumentamos n para que la campana de Gauss se vea más suave
mu, sigma = 0, 25  # Media y desviación estándar del ruido

# 2. Generación de datos
X = np.linspace(0, 100, n)
y_real = 2.0 * -X
ruido_gaussiano = np.random.normal(mu, sigma, n) # Ruido puro
y = y_real + ruido_gaussiano

# 3. Cálculo del residual
# En este caso ideal, el residual es exactamente el ruido que añadimos
residuales = y - y_real

# 4. Gráfico del Histograma
fig = plt.figure(figsize = (7, 5))
ax = plt.gca()

# Graficamos los residuales
ax.hist(residuales, bins = 30, color = "blue", edgecolor = "pink", alpha = 0.7)

ax.set_title("Distribución del Ruido Gaussiano (Residuales)")
ax.set_xlabel("Valor del Error")
ax.set_ylabel("Frecuencia")

# Añadimos una línea vertical en la media (0)
ax.axvline(0, color='red', linestyle='dashed', linewidth=1.5, label='Media (0)')
ax.legend()

ax.grid(True, axis = 'y', linestyle = "--", alpha = 0.3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm # Importamos para la curva teórica

# 1. Parámetros del ruido
n = 1000
mu = 0
sigma = 25  # Este es el "ruido" que queremos señalar

# 2. Generar datos y residuales
X = np.linspace(0, 100, n)
y_real = 2.0 * -X
ruido = np.random.normal(mu, sigma, n)
y = y_real + ruido
residuales = y - y_real

# 3. Gráfico
fig, ax = plt.subplots(figsize=(8, 5))

# Histograma normalizado (density=True para que coincida con la curva)
count, bins, ignored = ax.hist(residuales, bins=30, density=True,
                               alpha=0.6, color='seagreen', edgecolor='white', label='Residuales (Datos)')

# 4. SEÑALAR EL RUIDO GAUSSIANO (Curva Teórica)
x_teorica = np.linspace(min(bins), max(bins), 100)
p_teorica = norm.pdf(x_teorica, mu, sigma) # Función de densidad gaussiana

ax.plot(x_teorica, p_teorica, color='darkred', linewidth=3,
        label=f'Ruido Gaussiano Teórico ($\sigma$={sigma})')

# Formato adicional
ax.set_title("Validación del Ruido Gaussiano en los Residuales")
ax.set_xlabel("Valor del Residual")
ax.set_ylabel("Densidad de Probabilidad")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

```python
# orden de frecuencias
# un plot que tenga un bar dejar lo daros que sean nominales y presentalos en grafica de barras
# ademas agregar en cada barra la frecuancia
#2 grafica de barras ordinal con etiquetas y orden correcto
```

```python
#  PASO 1: DATOS NOMINALES (NUEVOS)
cats = ['Laptop', 'Tablet', 'Smartphone', 'Smartwatch']
conteos = np.array([80, 45, 120, 60])

#  PASO 2: ORDENAR POR FRECUENCIA
orden = np.argsort(conteos)[::-1]
cats_ordenados = np.array(cats)[orden]
conteos_ordenados = conteos[orden]

#  PASO 3: FIGURA Y EJE
fig = plt.figure(figsize=(6, 4))
ax = plt.gca()

#  PASO 4: GRÁFICA DE BARRAS
barras = ax.bar(cats_ordenados, conteos_ordenados, color="black")

#  PASO 5: AGREGAR FRECUENCIA EN CADA BARRA
for barra in barras:
    altura = barra.get_height()
    ax.text(barra.get_x() + barra.get_width()/2,
            altura,
            str(int(altura)),
            ha='center',
            va='bottom')

#  PASO 6: ETIQUETAS Y GRID
ax.set_title("Barras: Conteos por tipo de dispositivo")
ax.set_xlabel("Categoría")
ax.set_ylabel("Frecuencia")
ax.grid(True, axis='y', linestyle="--", alpha=0.4)

plt.show()
```

```python
#  PASO 1: DATOS ORDINALES
# Ejemplo: nivel de satisfacción (tiene orden lógico)
cats = ['Muy bajo', 'Bajo', 'Medio', 'Alto', 'Muy alto']
conteos = np.array([20, 35, 50, 40, 25])

# PASO 2: (NO se ordena automáticamente)
# En datos ordinales se respeta el orden lógico definido arriba

#  PASO 3: FIGURA Y EJE
fig = plt.figure(figsize=(6, 4))
ax = plt.gca()

#  PASO 4: GRÁFICA DE BARRAS
barras = ax.bar(cats, conteos, color="black")

#  PASO 5: AGREGAR FRECUENCIA EN CADA BARRA
for barra in barras:
    altura = barra.get_height()
    ax.text(barra.get_x() + barra.get_width()/2,
            altura,
            str(int(altura)),
            ha='center',
            va='bottom')

#  PASO 6: ETIQUETAS Y GRID
ax.set_title("Barras: Nivel de satisfacción (Datos ordinales)")
ax.set_xlabel("Categoría ordinal")
ax.set_ylabel("Frecuencia")
ax.grid(True, axis='y', linestyle="--", alpha=0.4)

plt.show()
```
