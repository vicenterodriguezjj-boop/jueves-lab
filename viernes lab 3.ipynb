### crear la variable

```python
import numpy as np
```

### crear el vector con sus elementos

```python
vector = np.array([1, 2, 3, 4, 5])
```

### depues extraemos cada una de las propiedades

### para la forma se utiliza shape

```python
print(f"Forma (shape): {vector.shape}")
```

### para el tamaño se utiliza size

```python
print(f"Tamaño (size): {vector.size}")
```

### para el tipo de objetos se utiliza type


```python
print(f"Tipo de objeto: {type(vector)}")
```

### para el tipo de contenido se utiliza dtype

```python
print(f"Tipo de contenido (dtype): {vector.dtype}")
```

#### despues se extraen los elementos solicitados

### extraer el elemento 2 y 3

```python
print(f"Elementos 2 y 3: {vector[1:3]}")
```

### extraer el primer elemento

```python
print(f"Primer elemento: {vector[0]}")
```

### extraer los ultimos 2 elementos

```python
print(f"Dos últimos elementos: {vector[-2:]}")
```

### ahora vamos con eel segundo vector

### primero se crea la matriz

```python
matriz = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(2, 4)
```

### extraer cada una de las propiedades que se ocupan

### la matriz

```python
print(f"Matriz:\n{matriz}")
```

### despues la froma que al igual se utiliza shape

```python
print(f"Forma (shape): {matriz.shape}")
```

### despues el tamaño donde se utiliza size

```python
print(f"Tamaño (size): {matriz.size}")
```

### tambien el tipo de contenido donde se utiliza dtype

```python
print(f"Tipo de contenido (dtype): {matriz.dtype}")
```

### despues extraer las mismas espesificaciones primer renglon

###

```python
print(f"Primer renglón completo: {matriz[0, :]}")
```

### primer elemento

```python
print(f"Primer elemento: {matriz[0, 0]}")
```

### ultimos dos elementos

```python
print(f"Dos últimos elementos (del total): {matriz.flatten()[-2:]}")
```
